```python
import json
from geojson import MultiPolygon, Polygon, Feature, FeatureCollection
import os
import pymongo
from pymongo import MongoClient
```

```python
#Funci√≥ per transformar un multipolygon a polygon
def multipolygon_to_polygon(feature):
    polygons = []
    properties = feature['properties']
    for coordinates in feature['geometry']['coordinates']:
        polygon = Polygon(coordinates)
        polygon_feature = Feature(geometry=polygon, properties=properties)
        polygons.append(polygon_feature)
    return polygons
```

```python
client = MongoClient("localhost:27017")
db = client["hsz"]
c = db['hsz']
"""SUbir todos los archivos polygon geojson a mongo"""

for filename in os.listdir('C:/Users/manya/Documents/Ironhack/Course/Final-project/geojson/poligon_geojson'): 
    if filename.endswith('.geojson'):
        file_path = os.path.join('C:/Users/manya/Documents/Ironhack/Course/Final-project/geojson/poligon_geojson', filename)

        # Load GeoJSON file
        with open(file_path) as file:
            data = json.load(file)

        # Insert each feature into the MongoDB collection
        for feature in data['features']:
            result = c.insert_one(feature)
            print(result.inserted_id)
```
