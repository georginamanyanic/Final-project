```python
# bucle
for filename in os.listdir('C:/Users/manya/Documents/Ironhack/Course/Final-project/geojson/old_geojson'): 
    if filename.endswith('.geojson'):
        file_path = os.path.join('C:/Users/manya/Documents/Ironhack/Course/Final-project/geojson/old_geojson', filename)
            # Load the GeoJSON file
        with open(file_path) as file:
            data = json.load(file)

        # Iterate over multipolygons
        for multipolygon in data['features']:
            if multipolygon['geometry']['type'] == 'MultiPolygon':
                for polygon in multipolygon['geometry']['coordinates']:
                
                    lista_puntos_un_poligono = list(set(tuple(map(tuple, obj)) for obj in polygon))

                    # Check if the initial point matches the end point
                    if lista_puntos_un_poligono[0][0] != lista_puntos_un_poligono[-1][-1]:
                        # Close the polygon by appending the first point at the end
                        lista_puntos_un_poligono.append(lista_puntos_un_poligono[0])
                    polygon = lista_puntos_un_poligono
        # Save the corrected GeoJSON
        with open(filename +'_updated.geojson', 'w') as file:
            json.dump(data, file)
```

```python
# bucle
for filename in os.listdir('C:/Users/manya/Documents/Ironhack/Course/Final-project/geojson/old_geojson'): 
    if filename.endswith('.geojson'):
        file_path = os.path.join('C:/Users/manya/Documents/Ironhack/Course/Final-project/geojson/old_geojson', filename)
            # Load the GeoJSON file
        with open(file_path) as file:
            data = json.load(file)

        # Iterate over multipolygons
        for multipolygon in data['features']:
            if multipolygon['geometry']['type'] == 'MultiPolygon':
                for polygon in multipolygon['geometry']['coordinates']:
                    unique_points = []
                    for point in polygon:
                        if point not in unique_points:
                            unique_points.append(point)

                    # Check if the initial point matches the end point
                    if unique_points[0] != unique_points[-1]:
                        # Close the polygon by appending the first point at the end
                        unique_points.append(unique_points[0])

                    polygon = unique_points


        # Save the corrected GeoJSON
        with open(filename +'_updated.geojson', 'w') as file:
            json.dump(data, file)
```
