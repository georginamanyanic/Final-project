```python
map = folium.Map(location=[40.2085,-3.713],zoom_start = 6)
color_hsz =  {
    '4b': '#FF0000',  # Rojo
    '5a': '#FFA500',  # Naranja
    '5b': '#FFFF00',  # Amarillo
    '6a': '#008000',  # Verde oscuro
    '6b': '#00FF00',  # Verde claro
    '7a': '#0000FF',  # Azul
    '7b': '#800080',  # Violeta
    '8a': '#FF00FF',  # Rosa
    '8b': '#FFC0CB',  # Rosa claro
    '9a': '#800000',  # Marrón oscuro
    '9b': '#FF4500',  # Naranja oscuro
    '10a': '#00FFFF',  # Cian
    '10b': '#808080',  # Gris
    '11a': '#FFD700',  # Oro
    '11b': '#800080',  # Púrpura
    '12a': '#008080'  # Verde azulado
}
def estilo_hsz(feature):
    hsz = feature['properties']['hzs']
    color = color_hsz.get(hsz, '#ffffff')  # Color blanco si el tipo de hsz no está definido en la lista
    return {
        'fillColor': color,
        'color': '#000000',
        'weight': 1,
        'fillOpacity': 0.5
    }
geojson_data = gpd.read_file('C:/Users/manya/Documents/Ironhack/Course/Final-project/geojson/tableau.geojson')

# Agregar el GeoJSON al mapa con el estilo personalizado
folium.GeoJson(geojson_data, style_function=estilo_hsz).add_to(map)

# Agregar la leyenda de colores al mapa
leyenda_colores_html = '<div style="position: fixed; top: 10px; right: 10px; background-color: white; padding: 10px; z-index: 1000;">'
leyenda_colores_html += '<h4>Hardiness Zones</h4>'

for hsz, color in color_hsz.items():
    leyenda_colores_html += f'<div style="margin-bottom: 5px;"><span style="background-color: {color}; width: 20px; height: 20px; display: inline-block;"></span> {hsz}</div>'

leyenda_colores_html += '</div>'

map.get_root().html.add_child(folium.Element(leyenda_colores_html))

#Agregar un marker de las coordenadas
folium.Marker(location = coordenadas, popup = 'Hardiness Zones: '+ hsz_).add_to(map)
```
