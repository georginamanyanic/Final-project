```python
from PIL import Image 
import os 
import imghdr
import random
import shutil

```

```python
def convert_png_to_jpg(folder_path):
    # Obtener la lista de archivos en la carpeta
    files = os.listdir(folder_path)

    for file in files:
        if file.endswith('.png'):
            # Crear la ruta completa del archivo PNG
            file_path = os.path.join(folder_path, file)

            # Abrir la imagen PNG y convertirla a formato JPG
            image = Image.open(file_path)
            image = image.convert('RGB')

            # Crear el nombre de archivo de salida en formato JPG
            filename = os.path.splitext(file)[0]
            output_path = os.path.join(folder_path, f"{filename}.jpg")

            # Guardar la imagen en formato JPG
            image.save(output_path, 'JPEG')

            # Eliminar el archivo PNG original
            os.remove(file_path)

    print('Conversión completa y archivos PNG eliminados.')
```

```python
convert_png_to_jpg('../images/Strawberry')
```

```python
def resize_images(original_folder): 
    
    for filename in os.listdir(original_folder): 
        file_path = f'{original_folder}/{filename}'
        try: 
            image = Image.open(file_path)
            image = image.resize((224, 224), Image.ANTIALIAS)
            image.save(file_path)
            print(f"Rescaled {filename} successfully.")
        except Exception as e:
                print(f"Error rescaling {filename}: {e}")
```

```python
resize_images('../images/Strawberry')
```

```python
def split_train_test_val (original_folder, name_vegetable):

    train_dir = f'../images/train/{name_vegetable}'
    test_dir = f'../images/test/{name_vegetable}'
    val_dir = f'../images/validation/{name_vegetable}'

    file_list = os.listdir(original_folder)

    # Obtener la cantidad de archivos según la proporción deseada
    num_train = int(len(file_list) * 0.7)
    num_test = int(len(file_list) * 0.15)
    num_validation = int(len(file_list) * 0.15)

    # Mezclar aleatoriamente la lista de archivos
    random.shuffle(file_list)

    for i, filename in enumerate(file_list):
        src_path = f'{original_folder}/{filename}'
        if i < num_train:
            dst_path = f'{train_dir}/{filename}'
        elif i < num_train + num_test:
            dst_path = f'{test_dir}/{filename}'
        else:
            dst_path = f'{val_dir}/{filename}'

        os.makedirs(os.path.dirname(dst_path), exist_ok=True) #crear dst_path 
        shutil.copy(src_path, dst_path) #copia el archivo de origen al de destino

    print('Distribución acabada')
```

```python
split_train_test_val ('../images/Strawberry','Strawberry')
```
